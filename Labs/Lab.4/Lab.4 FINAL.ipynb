1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        self.value = 0
        self.maximum = maximum

    def increment(self, n=1):
        if self.value + n > self.maximum:
            print("Error: counter cannot exceed maximum.")
        else:
            self.value += n

    def reset(self):
        self.value = 0
```

```python
## Test

c = Counter(3)
print("Initial:", c.value)

c.increment()
print("After +1:", c.value)

c.increment(2)
print("After +2:", c.value)

c.increment()                     
print("After overflow attempt:", c.value)

c.reset()
print("After reset:", c.value)    
```

```python
class PrivateCounter:
    def __init__(self, maximum):
        self.__value = 0
        self.__maximum = maximum

    def increment(self, n=1):
        if self.__value + n > self.__maximum:
            print("Error: counter cannot exceed maximum.")
        else:
            self.__value += n

    def reset(self):
        self.__value = 0

    def value(self):
        return self.__value

    def maximum(self):
        return self.__maximum

    def is_at_maximum(self):
        return self.__value == self.__maximum
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
class TriangleShape(Shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    def base(self):
        return self.__base

    def height(self):
        return self.__height

    def x(self):
        return self.__x

    def y(self):
        return self.__y

    def area(self):
        return 0.5 * self.__base * self.__height

    def perimeter(self):
        hyp = math.sqrt(self.__base ** 2 + self.__height ** 2)
        return self.__base + self.__height + hyp

    def get_points(self):
        x0, y0 = self.__x, self.__y
        x1, y1 = x0 + self.__base, y0
        x2, y2 = x0, y0 + self.__height
        points = [
            (x0, y0),
            (x1, y1),
            (x2, y2),
            ((x0 + x1) / 2, (y0 + y1) / 2),
            ((x1 + x2) / 2, (y1 + y2) / 2),
            ((x2 + x0) / 2, (y2 + y0) / 2),
        ]
        return points[:16]

    def contains_point(self, x, y):
        x0, y0 = self.__x, self.__y
        x1, y1 = x0 + self.__base, y0
        x2, y2 = x0, y0 + self.__height

        denom = (y1 - y2) * (x0 - x2) + (x2 - x1) * (y0 - y2)
        if denom == 0:
            return False
        a = ((y1 - y2) * (x - x2) + (x2 - x1) * (y - y2)) / denom
        b = ((y2 - y0) * (x - x2) + (x0 - x2) * (y - y2)) / denom
        c = 1 - a - b
        return (0 <= a <= 1) and (0 <= b <= 1) and (0 <= c <= 1)
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape, char='*'):
        self.shapes.append((shape, char))

    def remove_shape(self, index):
        if 0 <= index < len(self.shapes):
            self.shapes.pop(index)

    def paint(self, canvas):
        canvas.clear_canvas()
        for shape, char in self.shapes:
            paint_shape_on_canvas(shape, canvas, char=char)

    def __repr__(self):
        parts = []
        parts.append("RasterDrawing()")
        return "RasterDrawing()"  
```

```python
import math

class Shape:
    def get_points(self):
        raise NotImplementedError

    def contains_point(self, x, y):
        raise NotImplementedError

    def overlaps(self, other):
        for (x, y) in self.get_points():
            if other.contains_point(x, y):
                return True
        for (x, y) in other.get_points():
            if self.contains_point(x, y):
                return True
        return False
```

```python
class RectShape(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_points(self):
        pts = []
        steps = 4
        for i in range(steps):
            pts.append((self.__x + i*self.__length/(steps-1), self.__y))
        for i in range(steps):
            pts.append((self.__x + i*self.__length/(steps-1), self.__y + self.__width))
        for i in range(steps):
            pts.append((self.__x, self.__y + i*self.__width/(steps-1)))
        for i in range(steps):
            pts.append((self.__x + self.__length, self.__y + i*self.__width/(steps-1)))
        return pts[:16]

    def contains_point(self, x, y):
        return (self.__x <= x <= self.__x + self.__length and
                self.__y <= y <= self.__y + self.__width)

    def __repr__(self):
        return f"RectShape({self.__length}, {self.__width}, {self.__x}, {self.__y})"
```

```python
class CircleShape(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_points(self):
        pts = []
        for i in range(16):
            theta = 2*math.pi*i/16
            pts.append((self.__x + self.__radius*math.cos(theta),
                        self.__y + self.__radius*math.sin(theta)))
        return pts

    def contains_point(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        return dx*dx + dy*dy <= self.__radius*self.__radius

    def __repr__(self):
        return f"CircleShape({self.__radius}, {self.__x}, {self.__y})"
```

```python
class TriangleShape(Shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    def get_points(self):
        x0, y0 = self.__x, self.__y
        x1, y1 = x0 + self.__base, y0
        x2, y2 = x0, y0 + self.__height
        return [
            (x0, y0),
            (x1, y1),
            (x2, y2),
            ((x0+x1)/2, y0),
            (x0, (y0+y2)/2),
            ((x1+x2)/2, (y1+y2)/2)
        ][:16]

    def contains_point(self, x, y):
        x0, y0 = self.__x, self.__y
        x1, y1 = x0 + self.__base, y0
        x2, y2 = x0, y0 + self.__height

        denom = (y1-y2)*(x0-x2) + (x2-x1)*(y0-y2)
        if denom == 0:
            return False

        a = ((y1-y2)*(x-x2) + (x2-x1)*(y-y2)) / denom
        b = ((y2-y0)*(x-x2) + (x0-x2)*(y-y2)) / denom
        c = 1 - a - b

        return (0 <= a <= 1) and (0 <= b <= 1) and (0 <= c <= 1)

    def __repr__(self):
        return f"TriangleShape({self.__base}, {self.__height}, {self.__x}, {self.__y})"
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
def paint_shape_on_canvas(shape, canvas, char='*'):
    for (x, y) in shape.get_points():
        canvas.set_pixel(int(round(y)), int(round(x)), char)
```

```python
class RasterDrawing:
    def __init__(self, shapes_and_chars=None):
        self.shapes = []
        if shapes_and_chars:
            for shape, char in shapes_and_chars:
                self.shapes.append((shape, char))

    def add_shape(self, shape, char='*'):
        self.shapes.append((shape, char))

    def remove_shape(self, index):
        if 0 <= index < len(self.shapes):
            self.shapes.pop(index)

    def paint(self, canvas):
        canvas.clear_canvas()
        for shape, char in self.shapes:
            paint_shape_on_canvas(shape, canvas, char)

    def __repr__(self):
        shape_reprs = []
        for shape, char in self.shapes:
            shape_reprs.append(f"({repr(shape)}, {repr(char)})")
        return f"RasterDrawing([{', '.join(shape_reprs)}])"

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(repr(self))
```

```python
def raster_drawing_loader(filename):
    with open(filename, "r") as f:
        code = f.read()

    namespace = {
        "RasterDrawing": RasterDrawing,
        "RectShape": RectShape,
        "CircleShape": CircleShape,
        "TriangleShape": TriangleShape,
        "paint_shape_on_canvas": paint_shape_on_canvas,
        "Canvas": Canvas
    }

    return eval(code, namespace)
```
