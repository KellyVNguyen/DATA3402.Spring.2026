1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.value = 0
        self.max_value = max_value

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: counter cannot exceed maximum value.")

    def reset(self):
        self.value = 0

    def __str__(self):
        return f"Counter(value={self.value}, max={self.max_value})"
```

```python
# Test

# Test 1: Basic increment
c = Counter(3)
print(c)          # Counter(value=0, max=3)

c.increment()
c.increment()
print(c)          # Counter(value=2, max=3)

# Test 2: Increment past max
c.increment()     # value becomes 3
c.increment()     # should print error
print(c)          # Counter(value=3, max=3)

# Test 3: Reset
c.reset()
print(c)          # Counter(value=0, max=3)
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, max_value):
        self.__value = 0
        self.__max_value = max_value

    # --- Behavior Methods ---
    def increment(self):
        if self.__value < self.__max_value:
            self.__value += 1
        else:
            print("Error: counter cannot exceed maximum value.")

    def reset(self):
        self.__value = 0

    # --- Accessors / Getters ---
    def get_value(self):
        return self.__value

    def get_max_value(self):
        return self.__max_value

    def is_at_max(self):
        return self.__value == self.__max_value

    def __str__(self):
        return f"Counter(value={self.__value}, max={self.__max_value})"
```

```python
# Test

# Test 1: Basic increment
c = Counter(3)
print(c.get_value())      # 0
c.increment()
c.increment()
print(c.get_value())      # 2

# Test 2: Increment past max
c.increment()             # value becomes 3
c.increment()             # should print error
print(c.is_at_max())      # True

# Test 3: Reset
c.reset()
print(c.get_value())      # 0

# Test 4: Accessors
print(c.get_max_value())  # 3
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Rectangle:
    def __init__(self, x, y, length, width):
        # Private data members
        self.__x = x
        self.__y = y
        self.__length = length
        self.__width = width

    # --- Accessors / Getters ---
    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    # --- Geometry Methods ---
    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def __str__(self):
        return (f"Rectangle(x={self.__x}, y={self.__y}, "
                f"length={self.__length}, width={self.__width})")
```

```python
# Test

# Create a rectangle at (2, 3) with length 10 and width 5
r = Rectangle(2, 3, 10, 5)

# Test accessors
print(r.get_x())        # 2
print(r.get_y())        # 3
print(r.get_length())   # 10
print(r.get_width())    # 5

# Test area and perimeter
print(r.area())         # 50
print(r.perimeter())    # 30

# Test string representation
print(r)                # Rectangle(x=2, y=3, length=10, width=5)
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle:
    def __init__(self, x, y, radius):
        # Private data members
        self.__x = x
        self.__y = y
        self.__radius = radius

    # --- Accessors / Getters ---
    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_radius(self):
        return self.__radius

    # --- Geometry Methods ---
    def area(self):
        return math.pi * (self.__radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def __str__(self):
        return (f"Circle(center=({self.__x}, {self.__y}), "
                f"radius={self.__radius})")
```

```python
# Test

# Create a circle centered at (4, 5) with radius 3
c = Circle(4, 5, 3)

# Test accessors
print(c.get_x())        # 4
print(c.get_y())        # 5
print(c.get_radius())   # 3

# Test area and perimeter
print(c.area())         # 28.274333882308138
print(c.perimeter())    # 18.84955592153876

# Test string representation
print(c)                # Circle(center=(4, 5), radius=3)
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        # Private data members
        self.__x1 = x1
        self.__y1 = y1
        self.__x2 = x2
        self.__y2 = y2
        self.__x3 = x3
        self.__y3 = y3

    # --- Accessors ---
    def get_x(self):
        # Return the centroid x-coordinate (consistent with rectangle/circle having one x)
        return (self.__x1 + self.__x2 + self.__x3) / 3

    def get_y(self):
        # Return the centroid y-coordinate
        return (self.__y1 + self.__y2 + self.__y3) / 3

    def get_vertices(self):
        return ((self.__x1, self.__y1),
                (self.__x2, self.__y2),
                (self.__x3, self.__y3))

    # --- Helper: distance between two points ---
    def __dist(self, x1, y1, x2, y2):
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # --- Geometry ---
    def perimeter(self):
        a = self.__dist(self.__x1, self.__y1, self.__x2, self.__y2)
        b = self.__dist(self.__x2, self.__y2, self.__x3, self.__y3)
        c = self.__dist(self.__x3, self.__y3, self.__x1, self.__y1)
        return a + b + c

    def area(self):
        # Shoelace formula
        return abs(
            self.__x1*(self.__y2 - self.__y3) +
            self.__x2*(self.__y3 - self.__y1) +
            self.__x3*(self.__y1 - self.__y2)
        ) / 2

    def __str__(self):
        return (f"Triangle(({self.__x1}, {self.__y1}), "
                f"({self.__x2}, {self.__y2}), "
                f"({self.__x3}, {self.__y3}))")
```

```python
# Test

# A simple right triangle with vertices (0,0), (4,0), (0,3)
t = Triangle(0, 0, 4, 0, 0, 3)

print(t)                       # Triangle((0,0),(4,0),(0,3))
print("Centroid:", t.get_x(), t.get_y())
print("Area:", t.area())       # 6
print("Perimeter:", t.perimeter())  # 12
print("Vertices:", t.get_vertices())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
class Shape:
    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def perimeter_points(self, n=16):
        raise NotImplementedError("perimeter_points() must be implemented by subclasses")
```

```python
class Rectangle(Shape):
    def __init__(self, x, y, length, width):
        self.__x = x
        self.__y = y
        self.__length = length
        self.__width = width

    # Accessors
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_length(self): return self.__length
    def get_width(self): return self.__width

    # Geometry
    def area(self): return self.__length * self.__width
    def perimeter(self): return 2 * (self.__length + self.__width)

    # Perimeter points
    def perimeter_points(self, n=16):
        pts = []
        corners = [
            (self.__x, self.__y),
            (self.__x + self.__length, self.__y),
            (self.__x + self.__length, self.__y + self.__width),
            (self.__x, self.__y + self.__width)
        ]

        # 4 edges → divide n points across them
        points_per_edge = max(1, n // 4)

        for i in range(4):
            x1, y1 = corners[i]
            x2, y2 = corners[(i + 1) % 4]

            for k in range(points_per_edge):
                t = k / points_per_edge
                x = x1 + t * (x2 - x1)
                y = y1 + t * (y2 - y1)
                pts.append((x, y))

        return pts[:n]

    def __str__(self):
        return f"Rectangle(x={self.__x}, y={self.__y}, length={self.__length}, width={self.__width})"
```

```python
import math

class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    # Accessors
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    def get_radius(self): return self.__radius

    # Geometry
    def area(self): return math.pi * self.__radius**2
    def perimeter(self): return 2 * math.pi * self.__radius

    # Perimeter points
    def perimeter_points(self, n=16):
        pts = []
        for i in range(n):
            angle = 2 * math.pi * i / n
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            pts.append((x, y))
        return pts

    def __str__(self):
        return f"Circle(center=({self.__x}, {self.__y}), radius={self.__radius})"
```

```python
class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__x1 = x1; self.__y1 = y1
        self.__x2 = x2; self.__y2 = y2
        self.__x3 = x3; self.__y3 = y3

    def get_x(self):
        return (self.__x1 + self.__x2 + self.__x3) / 3

    def get_y(self):
        return (self.__y1 + self.__y2 + self.__y3) / 3

    def get_vertices(self):
        return ((self.__x1, self.__y1),
                (self.__x2, self.__y2),
                (self.__x3, self.__y3))

    def __dist(self, x1, y1, x2, y2):
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    def perimeter(self):
        a = self.__dist(self.__x1, self.__y1, self.__x2, self.__y2)
        b = self.__dist(self.__x2, self.__y2, self.__x3, self.__y3)
        c = self.__dist(self.__x3, self.__y3, self.__x1, self.__y1)
        return a + b + c

    def area(self):
        return abs(
            self.__x1*(self.__y2 - self.__y3) +
            self.__x2*(self.__y3 - self.__y1) +
            self.__x3*(self.__y1 - self.__y2)
        ) / 2

    def perimeter_points(self, n=16):
        pts = []
        vertices = self.get_vertices()
        points_per_edge = max(1, n // 3)

        for i in range(3):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % 3]

            for k in range(points_per_edge):
                t = k / points_per_edge
                x = x1 + t * (x2 - x1)
                y = y1 + t * (y2 - y1)
                pts.append((x, y))

        return pts[:n]

    def __str__(self):
        return f"Triangle({vertices})"
```

```python
# Test

r = Rectangle(0, 0, 10, 5)
print("Rectangle perimeter points:", r.perimeter_points())

c = Circle(0, 0, 5)
print("Circle perimeter points:", c.perimeter_points())

t = Triangle(0, 0, 4, 0, 0, 3)
print("Triangle perimeter points:", t.perimeter_points())
```

```python
class Shape:
    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def perimeter_points(self, n=16):
        raise NotImplementedError

    def contains_point(self, x, y):
        raise NotImplementedError("contains_point() must be implemented by subclasses")
```

```python
class Rectangle(Shape):
    def __init__(self, x, y, length, width):
        self.__x = x
        self.__y = y
        self.__length = length
        self.__width = width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def contains_point(self, x, y):
        return (self.__x <= x <= self.__x + self.__length and
                self.__y <= y <= self.__y + self.__width)
```

```python
import math

class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return math.pi * self.__radius**2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def contains_point(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        return dx*dx + dy*dy <= self.__radius * self.__radius
```

```python
class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__x1 = x1
        self.__y1 = y1
        self.__x2 = x2
        self.__y2 = y2
        self.__x3 = x3
        self.__y3 = y3

    def get_x(self):
        return (self.__x1 + self.__x2 + self.__x3) / 3

    def get_y(self):
        return (self.__y1 + self.__y2 + self.__y3) / 3

    def __area(self, x1, y1, x2, y2, x3, y3):
        return abs(
            x1*(y2 - y3) +
            x2*(y3 - y1) +
            x3*(y1 - y2)
        ) / 2

    def area(self):
        return self.__area(self.__x1, self.__y1,
                           self.__x2, self.__y2,
                           self.__x3, self.__y3)

    def contains_point(self, x, y):
        A = self.area()
        A1 = self.__area(x, y, self.__x2, self.__y2, self.__x3, self.__y3)
        A2 = self.__area(self.__x1, self.__y1, x, y, self.__x3, self.__y3)
        A3 = self.__area(self.__x1, self.__y1, self.__x2, self.__y2, x, y)

        return abs((A1 + A2 + A3) - A) < 1e-6
```

```python
# Rectangle test
r = Rectangle(0, 0, 10, 5)
print(r.contains_point(3, 2))   # True
print(r.contains_point(11, 2))  # False

# Circle test
c = Circle(0, 0, 5)
print(c.contains_point(3, 4))   # True (3-4-5 triangle)
print(c.contains_point(6, 0))   # False

# Triangle test
t = Triangle(0, 0, 4, 0, 0, 3)
print(t.contains_point(1, 1))   # True
print(t.contains_point(5, 5))   # False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def perimeter_points(self, n=16):
        raise NotImplementedError

    def contains_point(self, x, y):
        raise NotImplementedError

    def overlaps(self, other):
        # Check if any of my perimeter points are inside the other shape
        for (x, y) in self.perimeter_points():
            if other.contains_point(x, y):
                return True

        # Check if any of the other shape's perimeter points are inside me
        for (x, y) in other.perimeter_points():
            if self.contains_point(x, y):
                return True

        return False
```

```python
class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__x1 = x1
        self.__y1 = y1
        self.__x2 = x2
        self.__y2 = y2
        self.__x3 = x3
        self.__y3 = y3

    def get_x(self):
        return (self.__x1 + self.__x2 + self.__x3) / 3

    def get_y(self):
        return (self.__y1 + self.__y2 + self.__y3) / 3

    def __area(self, x1, y1, x2, y2, x3, y3):
        return abs(
            x1*(y2 - y3) +
            x2*(y3 - y1) +
            x3*(y1 - y2)
        ) / 2

    def area(self):
        return self.__area(self.__x1, self.__y1,
                           self.__x2, self.__y2,
                           self.__x3, self.__y3)

    def perimeter(self):
        def dist(x1, y1, x2, y2):
            return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

        a = dist(self.__x1, self.__y1, self.__x2, self.__y2)
        b = dist(self.__x2, self.__y2, self.__x3, self.__y3)
        c = dist(self.__x3, self.__y3, self.__x1, self.__y1)
        return a + b + c

    def perimeter_points(self, n=16):
        pts = []
        vertices = [
            (self.__x1, self.__y1),
            (self.__x2, self.__y2),
            (self.__x3, self.__y3)
        ]

        points_per_edge = max(1, n // 3)

        for i in range(3):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % 3]

            for k in range(points_per_edge):
                t = k / points_per_edge
                x = x1 + t * (x2 - x1)
                y = y1 + t * (y2 - y1)
                pts.append((x, y))

        return pts[:n]

    def contains_point(self, x, y):
        A = self.area()
        A1 = self.__area(x, y, self.__x2, self.__y2, self.__x3, self.__y3)
        A2 = self.__area(self.__x1, self.__y1, x, y, self.__x3, self.__y3)
        A3 = self.__area(self.__x1, self.__y1, self.__x2, self.__y2, x, y)

        return abs((A1 + A2 + A3) - A) < 1e-6
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Shape:
    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def perimeter_points(self, n=16):
        raise NotImplementedError

    def contains_point(self, x, y):
        raise NotImplementedError

    def overlaps(self, other):
        # Check if any of my perimeter points are inside the other shape
        for (x, y) in self.perimeter_points():
            if other.contains_point(x, y):
                return True

        # Check if any of the other shape's perimeter points are inside me
        for (x, y) in other.perimeter_points():
            if self.contains_point(x, y):
                return True

        return False
```

```python
# Rectangle

class Rectangle(Shape):
    def __init__(self, x, y, length, width):
        self.__x = x
        self.__y = y
        self.__length = length
        self.__width = width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)

    def perimeter_points(self, n=16):
        pts = []
        corners = [
            (self.__x, self.__y),
            (self.__x + self.__length, self.__y),
            (self.__x + self.__length, self.__y + self.__width),
            (self.__x, self.__y + self.__width)
        ]

        points_per_edge = max(1, n // 4)

        for i in range(4):
            x1, y1 = corners[i]
            x2, y2 = corners[(i + 1) % 4]

            for k in range(points_per_edge):
                t = k / points_per_edge
                x = x1 + t * (x2 - x1)
                y = y1 + t * (y2 - y1)
                pts.append((x, y))

        return pts[:n]

    def contains_point(self, x, y):
        return (self.__x <= x <= self.__x + self.__length and
                self.__y <= y <= self.__y + self.__width)

    def paint(self, canvas):
        for (x, y) in self.perimeter_points():
            canvas.set_pixel(int(x), int(y))
```

```python
# Circle

import math

class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return math.pi * self.__radius**2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def perimeter_points(self, n=16):
        pts = []
        for i in range(n):
            angle = 2 * math.pi * i / n
            x = self.__x + self.__radius * math.cos(angle)
            y = self.__y + self.__radius * math.sin(angle)
            pts.append((x, y))
        return pts

    def contains_point(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        return dx*dx + dy*dy <= self.__radius * self.__radius

    def paint(self, canvas):
        for (x, y) in self.perimeter_points():
            canvas.set_pixel(int(x), int(y))
```

```python
# Triangle

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__x1 = x1
        self.__y1 = y1
        self.__x2 = x2
        self.__y2 = y2
        self.__x3 = x3
        self.__y3 = y3

    def get_x(self):
        return (self.__x1 + self.__x2 + self.__x3) / 3

    def get_y(self):
        return (self.__y1 + self.__y2 + self.__y3) / 3

    def __area(self, x1, y1, x2, y2, x3, y3):
        return abs(
            x1*(y2 - y3) +
            x2*(y3 - y1) +
            x3*(y1 - y2)
        ) / 2

    def area(self):
        return self.__area(self.__x1, self.__y1,
                           self.__x2, self.__y2,
                           self.__x3, self.__y3)

    def perimeter(self):
        def dist(x1, y1, x2, y2):
            return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

        a = dist(self.__x1, self.__y1, self.__x2, self.__y2)
        b = dist(self.__x2, self.__y2, self.__x3, self.__y3)
        c = dist(self.__x3, self.__y3, self.__x1, self.__y1)
        return a + b + c

    def perimeter_points(self, n=16):
        pts = []
        vertices = [
            (self.__x1, self.__y1),
            (self.__x2, self.__y2),
            (self.__x3, self.__y3)
        ]

        points_per_edge = max(1, n // 3)

        for i in range(3):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % 3]

            for k in range(points_per_edge):
                t = k / points_per_edge
                x = x1 + t * (x2 - x1)
                y = y1 + t * (y2 - y1)
                pts.append((x, y))

        return pts[:n]

    def contains_point(self, x, y):
        A = self.area()
        A1 = self.__area(x, y, self.__x2, self.__y2, self.__x3, self.__y3)
        A2 = self.__area(self.__x1, self.__y1, x, y, self.__x3, self.__y3)
        A3 = self.__area(self.__x1, self.__y1, self.__x2, self.__y2, x, y)

        return abs((A1 + A2 + A3) - A) < 1e-6

    def paint(self, canvas):
        for (x, y) in self.perimeter_points():
            canvas.set_pixel(int(x), int(y))
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.pixels = [[' ' for _ in range(width)] for _ in range(height)]

    def set_pixel(self, x, y, char='*'):
        if 0 <= x < self.width and 0 <= y < self.height:
            self.pixels[y][x] = char

    def display(self):
        for row in self.pixels:
            print(''.join(row))
```

```python
class CompoundShape(Shape):
    def __init__(self):
        self.shapes = []

    def add(self, shape):
        self.shapes.append(shape)

    def paint(self, canvas):
        for s in self.shapes:
            s.paint(canvas)
```

```python
canvas = Canvas(40, 20)

r = Rectangle(2, 2, 10, 5)
c = Circle(20, 10, 6)
t = Triangle(30, 2, 35, 15, 25, 15)

shape = CompoundShape()
shape.add(r)
shape.add(c)
shape.add(t)

shape.paint(canvas)
canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []   # list of Shape objects

    def add(self, shape):
        self.shapes.append(shape)

    def remove(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def clear_canvas(self):
        self.canvas.pixels = [[' ' for _ in range(self.canvas.width)]
                              for _ in range(self.canvas.height)]

    def paint(self):
        # Clear the canvas
        self.clear_canvas()

        # Paint all shapes
        for s in self.shapes:
            s.paint(self.canvas)

        # Display the result
        self.canvas.display()
```

```python
# Create a canvas
canvas = Canvas(60, 20)

# Create a RasterDrawing
drawing = RasterDrawing(canvas)

# Create some shapes
r1 = Rectangle(5, 5, 10, 4)
c1 = Circle(25, 10, 6)
t1 = Triangle(40, 5, 50, 15, 30, 15)

# Add shapes to the drawing
drawing.add(r1)
drawing.add(c1)
drawing.add(t1)

print("Initial drawing:")
drawing.paint()

# Modify the drawing: remove one shape, add another
drawing.remove(c1)
r2 = Rectangle(20, 2, 15, 6)
drawing.add(r2)

print("\nModified drawing:")
drawing.paint()
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
