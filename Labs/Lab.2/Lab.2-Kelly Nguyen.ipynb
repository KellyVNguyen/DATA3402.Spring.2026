*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here

def game_status(board):
    n = len(board)

    for row in board:
        if row.count(1) == n:
            return 1
        if row.count(2) == n:
            return 2

    for col in range(n):
        column_values = [board[row][col] for row in range(n)]
        if column_values.count(1) == n:
            return 1
        if column_values.count(2) == n:
            return 2

    main_diag = [board[i][i] for i in range(n)]
    if main_diag.count(1) == n:
        return 1
    if main_diag.count(2) == n:
        return 2

    anti_diag = [board[i][n - 1 - i] for i in range(n)]
    if anti_diag.count(1) == n:
        return 1
    if anti_diag.count(2) == n:
        return 2

    for row in board:
        if 0 in row:
            return -1   

    return 0
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here

def parse_location(loc_str, n):
    if not isinstance(loc_str, str):
        return None
    if len(loc_str) != 2:
        return None

    row_char = loc_str[0].upper()
    col_char = loc_str[1]

    row_names = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")[:n]
    if row_char not in row_names:
        return None
    row = row_names.index(row_char)

    if not col_char.isdigit():
        return None
    col = int(col_char) - 1
    if col < 0 or col >= n:
        return None

    return (row, col)
````

````python
def place_mark_string(board, player, loc_str):
    n = len(board)
    loc = parse_location(loc_str, n)

    if loc is None:
        return False

    row, col = loc
    return place_mark(board, player, row, col)
````

````python
# Test your solution here

board = game_board(3)

print(place_mark_string(board, 1, "A1"))  
print(board)

print(place_mark_string(board, 2, "A1")) 
print(board)

print(place_mark_string(board, 2, "Z9")) 
print(board)

print(place_mark_string(board, 1, "B2"))  
print(board)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write your solution here

def play_ttt_game(n = 3):
    board = game_board(n)
    current_player = 1

    while True:
        draw_board_labeled(board)

        take_player_input(board, current_player)

        status = game_status(board)

        if status == 1:
            draw_board_labeled(board)
            print("Player 1 (X) wins!")
            return 1

        if status == 2:
            draw_board_labeled(board)
            print("Player 2 (O) wins!")
            return 2

        if status == 0:
            draw_board_labeled(board)
            print("It's a draw!")
            return 0

        current_player = 2 if current_player == 1 else 1
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
# Write your solution here

def get_empty_spaces(board):
    spaces = []
    n = len(board)
    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                spaces.append((r, c))
    return spaces

def would_win(board, player, r, c):
    board[r][c] = player
    result = game_status(board)
    board[r][c] = 0
    return result == player
````

````python
def computer_move(board, player):
    opponent = 1 if player == 2 else 2
    empty_spaces = get_empty_spaces(board)

    for (r, c) in empty_spaces:
        if would_win(board, player, r, c):
            place_mark(board, player, r, c)
            return True

    for (r, c) in empty_spaces:
        if would_win(board, opponent, r, c):
            place_mark(board, player, r, c)
            return True

    if empty_spaces:
        r, c = empty_spaces[0]
        place_mark(board, player, r, c)
        return True

    return False
````

````python
def play_ttt_game_vs_computer(n = 3):
    board = game_board(n)
    human = 1
    computer = 2
    current = human

    while True:
        draw_board_labeled(board)

        if current == human:
            take_player_input(board, human)
        else:
            print("Computer is thinking...")
            computer_move(board, computer)

        status = game_status(board)

        if status == human:
            draw_board_labeled(board)
            print("You win!")
            return
        if status == computer:
            draw_board_labeled(board)
            print("Computer wins!")
            return
        if status == 0:
            draw_board_labeled(board)
            print("It's a draw!")
            return

        current = computer if current == human else human
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here

def evaluate_position(board, computer, human):
    status = game_status(board)
    if status == computer:
        return 1
    if status == human:
        return -1
    if status == 0:   
        return 0
    return None        
````

````python
def search_best_move(board, player, computer, human, depth, max_depth):
    score = evaluate_position(board, computer, human)
    if score is not None:
        return score, None
    if depth == max_depth:
        return 0, None

    empty = get_empty_spaces(board)

    if player == computer:
        best_score = -float("inf")
        best_move = None

        for (r, c) in empty:
            board[r][c] = computer
            score, _ = search_best_move(board, human, computer, human, depth+1, max_depth)
            board[r][c] = 0

            if score > best_score:
                best_score = score
                best_move = (r, c)

        return best_score, best_move

    else:
        best_score = float("inf")
        best_move = None

        for (r, c) in empty:
            board[r][c] = human
            score, _ = search_best_move(board, computer, computer, human, depth+1, max_depth)
            board[r][c] = 0

            if score < best_score:
                best_score = score
                best_move = (r, c)

        return best_score, best_move
````

````python
def computer_move_exhaustive(board, computer, max_depth = 4):
    human = 1 if computer == 2 else 2
    _, move = search_best_move(board, computer, computer, human, 0, max_depth)

    if move is None:
        return False

    r, c = move
    place_mark(board, computer, r, c)
    return True
````

````python
def play_ttt_game_vs_smart_computer(n = 3, max_depth = 4):
    board = game_board(n)
    human = 1
    computer = 2
    current = human

    while True:
        draw_board_labeled(board)

        if current == human:
            take_player_input(board, human)
        else:
            print("Computer is thinking...")
            computer_move_exhaustive(board, computer, max_depth)

        status = game_status(board)

        if status == human:
            draw_board_labeled(board)
            print("You win!")
            return
        if status == computer:
            draw_board_labeled(board)
            print("Computer wins!")
            return
        if status == 0:
            draw_board_labeled(board)
            print("It's a draw!")
            return

        current = computer if current == human else human
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here

def play_smart_vs_simple(n, max_depth_smart = 4):
    board = game_board(n)
    smart = 1      
    simple = 2     
    current = smart

    while True:
        if current == smart:
            computer_move_exhaustive(board, smart, max_depth_smart)
        else:
            computer_move(board, simple)

        status = game_status(board)

        if status == smart:
            return 1
        if status == simple:
            return 2
        if status == 0:
            return 0

        current = simple if current == smart else smart
````

````python
def measure_smart_win_rate(n, games = 10, max_depth_smart = 4):
    smart_wins = 0
    simple_wins = 0
    draws = 0

    for _ in range(games):
        result = play_smart_vs_simple(n, max_depth_smart)
        if result == 1:
            smart_wins += 1
        elif result == 2:
            simple_wins += 1
        else:
            draws += 1

    print(f"{n}x{n} grid over {games} games:")
    print(f"  Smart wins:  {smart_wins}")
    print(f"  Simple wins: {simple_wins}")
    print(f"  Draws:       {draws}")
    print(f"  Smart win rate: {smart_wins / games:.2f}")
````

````python
# Test your solution here

measure_smart_win_rate(3, games = 10, max_depth_smart = 6)
measure_smart_win_rate(4, games = 10, max_depth_smart = 4)
measure_smart_win_rate(5, games = 10, max_depth_smart = 3)
````
